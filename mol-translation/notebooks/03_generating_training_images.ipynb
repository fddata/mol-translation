````python
from utils.draw_helpers import generate_labelled_images_from_inchi
from utils.inchi_helpers import get_inchi_strings_for_atom

import pandas as pd
````

We want to generate a list of inchis that meet the molecule counts that we require. 

The following creates a dict of inchis for the given elements.

````python
%%time
# df_train_labels = pd.read_csv("../data/train_labels.csv", nrows=1000)
df_train_labels = pd.read_csv("../data/train_labels.csv")

df_formula = pd.DataFrame()
df_formula["InChI"] = df_train_labels["InChI"]
df_formula["formula"] = df_formula["InChI"].str.split("/", n=2, expand=True)[1]
````

````python
%%time
elements = ["B", "Br", "C","Cl","F","I","N","O","P","S","Si"]
elements_mini = ["S", "Si", "C"]

inchi_map = {}
TARGET_COUNT = 20_000

for e in elements:
    # only want rows that have the element we want (speeds up sampling)
    mask = df_formula["formula"].str.contains(fr"{e}(?![a-z])(\d*)", regex=True)
    inchi_map[e] =  get_inchi_strings_for_atom(e, TARGET_COUNT, df_formula[mask])
````

The next step is to generate the required number of images for each element in the `inchi_map` dictionary

````python
%%time
for e in elements:

    if e == "H":
        continue

    generate_labelled_images_from_inchi(
        inchi_list=inchi_map[e],
        height=350,
        width=350,
        box_size=18,
        target_atom=e,
        target_image_count=TARGET_COUNT,
        save_path_prefix=f'../data/images/train_synthetic/'
    )
````
