# System setup - checking GPU and Torch

First we want to see if our system can detect a GPU:

```python
!nvidia-smi
```

This output will vary system to system, but for mine:
- NVIDIA driver is installed (535.183.01)
- A GPU is detected (RTX 3060) 👴
- A CUDA runtime is available (12.2)
- The GPU is currently in use (17W / 1.6 Gb)

We still need to know if Pytorch can find the GPU:

```python
import torch
```

```python
if torch.cuda.is_available():
    print("PyTorch version:", torch.__version__)
    print("CUDA available:", torch.cuda.is_available())
    print("CUDA version (PyTorch build):", torch.version.cuda)
    print("GPU count:", torch.cuda.device_count())
    
    for i in range(torch.cuda.device_count()):
        print(f"GPU name: {i+1}: {torch.cuda.get_device_name(i)}")
```

Fantastic! Everything we need for YOLO training is available.
